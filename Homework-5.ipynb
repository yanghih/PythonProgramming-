>>> print c == 17

>>> print d != 9

```python
import math

class Counter():
    
    #1번
    def __init__(self, count, step = 1):
        self.count = count
        self.step = step
    
    #2번
    def __str__(self):
        return '[Count (step : ' + str(self.step) + ')] ' + str(self.count)
    
    #3번
    def incr(self):
        self.count += self.step
        return self
    
    #4번
    def __call__(self):
        return self.incr()
    
    #5번
    def __add__(self, other):
        self.count += other
        return self
    
    def __sub__(self, other):
        self.count -= other
        return self
    
    #6번
    def __lt__(self, y): # <
        return self.count < y
    def __gt__(self, y): #>
        return self.count > y
    def __eq__(self, y): #==
        return self.count == y
    def __ne__(self, y): #==
        return self.count != y    
    
        
    
#1번
c = Counter(10) 
d = Counter(10, 2)

#2번
print c
print d

#3번
c.incr()
d.incr()

print c
print d

#4번
c()
d()
print c
print d

#5번
c = c + 5 
d = d - 5
print c
print d

#6번
print c > 10
print d > 10
print c < 10
print d < 10
print c == 17
print d != 9
```

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
```

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result
    


    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    def __or__(self, B): # 합집합 메소드 이다.
        self = s
        q = []
        for x in self:
            q.append(x)
        for x in B:
            q.append(x)
        # 집합 2개를 q에 다 넣는다.
        q.sort() #정렬해준다.
        return MySet(q)
        #Myset(q)를 불러주고 중복된 숫자가 없도록 만든다.
    
    def __and__(self, B):
        self = s
        q = []
        for x in self:
            for y in B:
                if (x == y):
                    q.append(y)
                    break
        # x, y를 비교해서 같은 숫자가 있다면 q에 넣는다.
        return MySet(q)
        #Myset(q)를 불러주고 중복된 숫자가 없도록 만든다.
    
    def __sub__(self, B):
        self = s
        q = []
        for x in self:
            if((x in B) == False):
                q.append(x)
        # 첫번쨰 집합에 있는 원소들이 2번째 집합에 존재하지 않으면 q에 넣는다.
        return MySet(q)
        #Myset(q)를 불러주고 중복된 숫자가 없도록 만든다.
    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
    
    #1번
    u = s | t
    print u
    
    #2번
    u = s & t 
    print u
    
    #3번
    s = MySet([1, 2, 3]) 
    t = MySet([3, 4, 5])
    u = s - t
    print u
```

100! 의 자리수를 모두 더하면 얼마입니까?

```python
def myfact(num):
    if (num == 0):
        return 1
    elif (num == 1):
        return 1
    else:
        return num * myfact(num-1)

num = myfact(100)

sum = 0
num = str(num)
for x in range(len(num)):    
    sum += int(num[x])
print sum
```

### 프로젝트 오일러 21) n의 약수들 중에서 자신을 제외한 것의 합을 d(n)으로 정의했을 때, 서로 다른 두 정수 a, b에 대하여 d(a) = b 이고 d(b) = a 이면 

a, b는 친화쌍이라 하고 a와 b를 각각 친화수(우애수)라고 합니다.

예를 들어 220의 약수는 자신을 제외하면 1, 2, 4, 5, 10, 11, 20, 22, 44, 55, 110 이므로 그 합은 d(220) = 284 입니다.
또 284의 약수는 자신을 제외하면 1, 2, 4, 71, 142 이므로 d(284) = 220 입니다.
따라서 220과 284는 친화쌍이 됩니다.

10000 이하의 친화수들을 모두 찾아서 그 합을 구하세요.

```python
x = 2

final = []
while(x<= 10000):
    l = []
    l2 = []
    sum1 = 0
    sum2 = 0
    for i in range(1, x):
        if(x % i ==  0):
            l.append(i)
    for j in l:
        sum1 += j
    for k in range(1,sum1):
        if(sum1 % k == 0):
            l2.append(k)
    for jj in l2:
        sum2 += jj
    if(x != sum1):
        if(x == sum2):
            final.append(x)    
    x += 1
    
print final

sum = 0
for y in final:
    sum += y
print sum
```
