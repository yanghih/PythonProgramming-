### 문제5. 다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.
  - 1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
        >>> addall([1])
        1
        >>> addall([1, 2, 3, 4, 5, 6, 7, 8, 9])
        45
  - 2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
        >>> addallodd([1])
        1
        >>> addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])
        25
  - [참고]: 리스트 내포 방식으로 addall() 및 addallodd()함수를 만들 때엔 리스트 내포에 의해 생성된 리스트 결과에 대해 내장 함수를 사용하는 방식을 고려해보시오 (주의: 리스트 내포 방식의 addall() 및 addallodd() 함수의 몸체는 단 1줄로 작성되어야 한다.)

```python
# 1-1) for~in 리터럴
def addall(L):
    a = 0
    for x in L:
        a = a + x
    return a
    
print addall([1])
print addall([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

```python
# 1-2) 리스트 내포방식
def addall(L):
    a = sum([x for x in L])
    return a
    
print addall([1])
print addall([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

```python
# 2-1) for~in 리터럴
def addallodd(L):
    a = 0
    for x in L:
        if(x % 2 == 0) :
            a = a
        else :
            a = a + x
    return a
    
print addallodd([1])
print addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

```python
# 2-2) 리스트 내포방식
def addallodd(L):
    a = sum([x for x in L if( x % 2 != 0)])
    return a
    
print addallodd([1])
print addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

  - [참고]: d.keys(), d.values()를 통해 리스트를 얻어낸 후 리스트가 지니고 있는 sort(cmp=None, key=None, reverse=False)함수를 활용하시오.

  - 1) 키의 알파벳 오름차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.

```python
d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}
l = []

#1)
for x in d.keys() :
    for y in d.values():
        if (d[x] == y) :
            l.append((x, y))
            
l.sort()
print l

#2)
l.sort(reverse=True)
print l

#3)

l = []
a = d.values()
a.sort()

for y in a :
    for x in d.keys():
        if (d[x] == y) :
            l.append((x, y))
print l


#4)
l = []
a = d.values()
a.sort(reverse = True)

for y in a :
    for x in d.keys():
        if (d[x] == y) :
            l.append((x, y))
print l
```

### 프로젝트 오일러 9)
  - 세 자연수 a, b, c 가 피타고라스 정리 a^2 + b^2 = c^2 를 만족하면 피타고라스 수라고 부릅니다 (여기서 a < b < c ).
  - 예를 들면 3^2 + 4^2 = 9 + 16 = 25 = 5^2이므로 3, 4, 5는 피타고라스 수입니다.

  - a + b + c = 1000 인 피타고라스 수 a, b, c는 한 가지 뿐입니다. 이 때, a × b × c 는 얼마입니까?

```python
for a in range(3, 1000) :
    for b in range(a+1, 1000) :
        for c in range(b+1, 1000) :
            sum= a**2 + b**2
            if (sum == c**2):
                total = a + b +c
                if(total == 1000) :
                    print a * b * c                    
```

    예: ['world', 'hello', 'python', 'world', '컴퓨터공학부', '지향'] ---> {'world': 2, 'hello': 1, 'python': 1, '컴퓨터공학부': 1, '지향': 1}


  - 요구사항 4. 다음 3개의 사이트에 대한 결과를 출력하시오.
        http://www.naver.com
        http://www.daum.net
        http://www.nytimes.com

```python
##########################컴퓨터공학부 홈페이지
import urllib2
import string

source = urllib2.urlopen("http://cse.kut.ac.kr/").read()

#print source # 추후 숙제 답안 제출 때엔 왼쪽 라인은 주석처리할 것

for i in range(0, source.count('<!--')):
    if source.find("<!--") != -1:
        source = source.replace(source[source.find("<!--"):source.find("-->")+3:],"")
#print source

for i in range(0,source.count("<")):
    if source.find("<") != -1:
        source = source.replace(source[source.find("<"):source.find(">")+1:],"")
#print source   # 순수 단어 출력

words = source.split() # 단어 나누기
l_punc = [] # 구두문자 리스트
l_word= [] # 단어들 리스트로(구두문자 없앤)
dic = {} # 사전 형식으로

s_punc = string.punctuation

for x in s_punc:
    l_punc.append(x)
l_punc.append('···')
l_punc.append('”')


for word in words:
    for y in l_punc:
        word = word.replace(y, '')
    if (word != ''):
        l_word.append(word)

for i in l_word:
    cou = l_word.count(i)
    dic[i] = cou
for dic_key in dic:
    print str(dic_key)+ " : "+str(dic[dic_key])
```
